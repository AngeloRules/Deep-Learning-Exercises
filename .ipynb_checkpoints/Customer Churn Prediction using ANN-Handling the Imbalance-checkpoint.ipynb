```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
customer_data = pd.read_csv('/Users/user/Desktop/Deep Learning Exercises/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
customer_data.info()
```

```python
customer_data = customer_data.drop('customerID',axis=1)
```

```python
sns.pairplot(customer_data)
```

```python
pd.to_numeric(customer_data['TotalCharges'],errors='coerce')
new_customer_data = customer_data[customer_data['TotalCharges']!=' ']
```

```python
new_customer_data
```

```python
new_customer_data['TotalCharges'] = new_customer_data['TotalCharges'].apply(float)
```

```python
new_customer_data.info()
```

```python
sns.histplot(data=new_customer_data,x='tenure',hue='Churn')
```

```python
sns.histplot(data=new_customer_data,x='MonthlyCharges',hue='Churn')
```

```python
for column in new_customer_data.columns:
    if new_customer_data[column].dtypes == 'object':
        print(f'{column} ==== ',new_customer_data[column].unique())
```

```python
new_customer_data.replace('No internet service','No',inplace=True)
new_customer_data.replace('No phone service','No',inplace=True)
```

```python
for column in new_customer_data.columns:
    if new_customer_data[column].dtypes == 'object':
        print(f'{column} ==== ',new_customer_data[column].unique())
```

```python
for column in new_customer_data.columns:
    if 'Yes' in new_customer_data[column].unique():
        new_customer_data[column].replace('Yes',1,inplace=True)
        new_customer_data[column].replace('No',0,inplace=True)
        print(f'{column} ==== ',new_customer_data[column].unique())
```

```python
new_customer_data['gender'].replace({'Male':1,'Female':0},inplace=True)
```

```python
for column in new_customer_data.columns:
    print(f'{column} ==== ',new_customer_data[column].unique())
```

```python
new_data_cats = pd.get_dummies(data=new_customer_data,columns=['InternetService','Contract','PaymentMethod'])
new_data_cats.info()
```

```python
new_data_cats[['tenure','MonthlyCharges','TotalCharges']] = MinMaxScaler().fit_transform(new_data_cats[['tenure','MonthlyCharges','TotalCharges']])
new_data_cats.head()
```

```python
x = new_data_cats.drop(['Churn'],axis=1)
y = new_data_cats['Churn']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = tf.keras.Sequential([
    keras.layers.Dense(26,input_shape=(26,),activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')  
])

call_back = tf.keras.callbacks.TensorBoard(log_dir='logs/Churn_ANN',histogram_freq=1)
model.compile(
    metrics=['accuracy'],
    optimizer = 'adam',
    loss = 'binary_crossentropy'
)
model.fit(x_train,y_train,epochs=1000,callbacks=call_back)
```

```python
model.evaluate(x_test,y_test)
```

```python
predictions = model.predict(x_test)
```

```python
prediction = [1 if i > 0.5 else 0 for i in predictions]
```

```python
print(classification_report(y_test,prediction))
```

```python
confusion_matrix = tf.math.confusion_matrix(labels=y_test,predictions=prediction)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(confusion_matrix,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfr = RandomForestClassifier()
rfr.fit(x_train,y_train)
prediction = rfr.predict(x_test)
print(classification_report(y_test,prediction))
```

```python
def ANN(x_train,y_train,x_test,y_test,weights):
    model = keras.Sequential([
        keras.layers.Dense(26,input_dim=26,activation='relu'),
        keras.layers.Dense(20,activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')
    ])
    model.compile(
    metrics=['accuracy'],
    optimizer='adam',
        loss='binary_crossentropy'
    )
    if weights == -1:
        model.fit(x_train,y_train,epochs=100)
    else:
        model.fit(x_train,y_train,epochs=100,class_weights=weights)
    print(model.evaluate(x_test,y_test))
    predictions = model.predict(x_test)
    predictions = np.round(predictions)
    
    print(classification_report(y_test,predictions))
```

**Without any data imbalance management techniques**

```python
ANN(x_train,y_train,x_test,y_test,weights=-1)
```

**Undersampling the majority class**

```python
churn_class, not_churn_class = new_data_cats[new_data_cats['Churn']==1],new_data_cats[new_data_cats['Churn']==0]
```

```python
not_churn_class.shape
```

```python
churn_class.shape
```

```python
new_not_churn_class = not_churn_class.sample(1869)
```

```python
undersampled_data = pd.concat([new_not_churn_class,churn_class],axis=0)
```

```python
undersampled_data['Churn'].value_counts()
```

```python
x = undersampled_data.drop('Churn',axis=1)
y = undersampled_data['Churn'] 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
ANN(x_train,y_train,x_test,y_test,weights=-1)
```

**OverSampling the minority class**

```python
new_churn = churn_class.sample(5163,replace=True)
```

```python
over_sampled_data = pd.concat([new_churn,not_churn_class],axis=0)
over_sampled_data['Churn'].value_counts()
```

```python
x = over_sampled_data.drop('Churn',axis=1)
y = over_sampled_data['Churn'] 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
ANN(x_train,y_train,x_test,y_test,weights=-1)
```

**Using SMOTE (Synthetic Minority Oversampling TEchnique)**

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy='minority')
```

```python
x = new_data_cats.drop('Churn',axis=1)
y = new_data_cats['Churn']
```

```python
x_smote,y_smote = smote.fit_resample(x,y)
```

```python
y_smote.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_smote,y_smote,test_size=0.2)
```

```python
ANN(x_train,y_train,x_test,y_test,weights=-1)
```
