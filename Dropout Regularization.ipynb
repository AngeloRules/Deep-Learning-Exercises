```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import tensorflow as tf
from tensorflow import keras
```

```python
sonar_data = pd.read_csv('/Users/user/Desktop/Deep Learning Exercises/sonar.csv',header=None)
```

```python
sonar_data
```

```python
x = sonar_data.drop(60,axis=1)
y = sonar_data[60]
```

```python
y = y.apply(lambda x : 1 if x == 'R' else 0 )
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

logs = tf.keras.callbacks.TensorBoard(log_dir='logs/WithoutDropout',histogram_freq=1)

model.compile(
    metrics=['accuracy'],
    loss='binary_crossentropy',
    optimizer = 'adam'
)

model.fit(x_train,y_train,epochs=100,callbacks=logs,batch_size=8)
```

```python
model.evaluate(x_test,y_test)
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])

logs = tf.keras.callbacks.TensorBoard(log_dir='logs/WithDropout',histogram_freq=1)

model.compile(
    metrics=['accuracy'],
    loss='binary_crossentropy',
    optimizer = 'adam'
)

model.fit(x_train,y_train,epochs=100,callbacks=logs,batch_size=8)
```

```python
model.evaluate(x_test,y_test)
```
